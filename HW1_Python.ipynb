## Problem 2: Random Variables, Expectation and Variance

In this problem, you will simulate observations of random variables from several common distributions.

To do: 
For each distribution (Poisson and Exponential)
1. generate $n=1000$ observations
2. visually compare the observed (empirical) pdf with the theoretical pdf of the random variable
3. estimate the expected value
4. estimate the variance and standard deviation
5. repeat steps 1-4 for $n=30$ observations
    
Examples with the Normal and Binomial are given below.

```python
### The Normal (Gaussian) Distribution
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

import numpy as np
import math

# 1000 observations 
X = norm.rvs(size=1000,loc=10,scale=1)

# empirical pdf
ax.hist(X, density=True, histtype='stepfilled', alpha=0.2)

# theoretical pdf in red
xinput=np.linspace(7, 13, 50)  # inputs to plug into pdf formula
ax.plot(xinput, norm.pdf(xinput, loc=10, scale=1),'r-', lw=4, alpha=0.6, label='norm pdf')
plt.show()


# expected value is approximately the average of lots of observations of X 
print(np.mean(X))

# variance is approximately
print(np.mean((X-np.mean(X))**2))
# which is the same as
print(np.mean(X**2)-(np.mean(X))**2)
# or we could use 
print(np.var(X))

# standard deviation is approximately
print(math.sqrt(np.mean((X-np.mean(X))**2)))
# or we could use
print(np.std(X))
```

```python
### The Normal (Gaussian) Distribution

# 30 observations 
X = norm.rvs(size=30,loc=10,scale=1)

# empirical pdf
fig, ax = plt.subplots(1, 1)
ax.hist(X, density=True, histtype='stepfilled', alpha=0.2)

# theoretical pdf in red
xinput=np.linspace(7, 13, 50)  # inputs to plug into pdf formulaepfilled'
ax.plot(xinput, norm.pdf(xinput, loc=10, scale=1),'r-', lw=4, alpha=0.6, label='norm pdf')
plt.show()


# expected value is approximately the average of lots of observations of X 
print("The expected value as estimated by the mean of 30 observations is ",np.mean(X))

# variance is approximately 
print("The variance as estimated with 30 observations is ",np.var(X))

# standard deviation is approximately
print("The standard deviation as estimated with 30 observations is ", np.std(X))
```

```python
### The Binomial Distribution
from scipy.stats import binom
import pandas as pd

# 1000 observations 
X = binom.rvs(size=1000, n=10,p=0.2)

# empirical pdf
fig, ax = plt.subplots(1, 1)
countsX = pd.crosstab(X, columns="counts")
ax.bar(np.linspace(min(X),max(X), (max(X)+1)), countsX.counts/1000, alpha=0.2)

# theoretical pdf in red
xinput=np.linspace(0, 10, 11)  # inputs to plug into pdf formula
ax.vlines(xinput,0, binom.pmf(xinput, n=10, p=0.2), lw=4, color="red", alpha=0.6, label='Binomial pdf')
plt.show()


# expected value is approximately the average of lots of observations of X 
print("The expected value as estimated by the mean of 1000 observations is ",np.mean(X))

# variance is approximately 
print("The variance as estimated with 1000 observations is ",np.var(X))

# standard deviation is approximately
print("The standard deviation as estimated with 1000 observations is ", math.sqrt(np.var(X)))
```

```python
### The Binomial Distribution
from scipy.stats import binom
import pandas as pd

# 30 observations 
X = binom.rvs(size=30, n=10,p=0.2)

# empirical pdf
fig, ax = plt.subplots(1, 1)
countsX = pd.crosstab(X, columns="counts")
ax.bar(np.linspace(min(X),max(X), (max(X)+1)), countsX.counts/30, alpha=0.2)

# theoretical pdf in red
xinput=np.linspace(0, 10, 11)  # inputs to plug into pdf formula (notice pdf for discrete is a "pmf")
ax.vlines(xinput,0, binom.pmf(xinput, n=10, p=0.2), lw=4, color="red", alpha=0.6, label='Binomial pdf')
plt.show()


# expected value is approximately the average of lots of observations of X 
print("The expected value as estimated by the mean of 30 observations is ",np.mean(X))

# variance is approximately 
print("The variance as estimated with 30 observations is ",np.var(X))

# standard deviation is approximately
print("The standard deviation as estimated with 30 observations is ", math.sqrt(np.var(X)))
```
